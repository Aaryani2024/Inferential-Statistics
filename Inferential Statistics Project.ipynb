#### The breaking strength of gunny bags used for packaging cement is normally distributed with a mean of 5 kg per sq. centimeter and a standard deviation of 1.5 kg per sq. centimeter. The quality team of the cement company wants to know the following about the packaging material to better understand wastage or pilferage within the supply chain; Answer the questions below based on the given information; (Provide an appropriate visual representation of your answers, without which marks will be deducted)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from scipy.stats import norm
import os
```

#### 2.1 What proportion of the gunny bags have a breaking strength less than 3.17 kg per sq cm?

```python
# joining all subplots
fig, ax=plt.subplots()

mean=5
std_dev=1.5

# Getting observations for distribution
x=np.linspace(mean-3*std_dev,mean+3*std_dev,100)
y=norm.pdf(x,loc=5,scale=1.5)

# Plotting the distribution with lables
plt.plot(x,y)
plt.xlabel('Breaking Strength')
plt.ylabel('probability Density')
plt.title('Normal Distribution with mean 5 and standard deviation 1.5')

# Putting a vertical line & filling the area
plt.axvline(3.17,c='r',linestyle='--')
ax.fill_between(x,y,where=(x<3.17),alpha=0.5,color='green')

# show plot & proportion
proportion=1-stats.norm.cdf(3.17,loc=5,scale=1.5)
plt.show()
```

#### 2.2 What proportion of the gunny bags have a breaking strength at least 3.6 kg per sq cm.?

```python
# joining all subplots
fig, ax=plt.subplots()

mean=5
std_dev=1.5

# Getting observations for distribution
x=np.linspace(mean-3*std_dev,mean+3*std_dev,100)
y=norm.pdf(x,loc=5,scale=1.5)

# Plotting the distribution with lables
plt.plot(x,y)
plt.xlabel('Breaking Strength')
plt.ylabel('probability Density')
plt.title('Normal Distribution with mean 5 and standard deviation 1.5')

# Putting a vertical line & filling the area
plt.axvline(3.6,c='r',linestyle='--')
ax.fill_between(x,y,where=(x<=3.6),alpha=0.5,color='green')

# show plot & proportion
proportion=1-stats.norm.cdf(3.6,loc=5,scale=1.5)
plt.show()
```

#### 2.3 What proportion of the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm.?

```python
# joining all subplots
fig, ax=plt.subplots()

mean=5
std_dev=1.5

# Getting observations for distribution
x=np.linspace(mean-3*std_dev,mean+3*std_dev,100)
y=norm.pdf(x,loc=5,scale=1.5)

# Plotting the distribution with lables
plt.plot(x,y)
plt.xlabel('Breaking Strength')
plt.ylabel('probability Density')
plt.title('Normal Distribution with mean 5 and standard deviation 1.5')

# Putting a vertical line & filling the area
plt.axvline(5,c='r',linestyle='--')
plt.axvline(5.5,c='r',linestyle='--')
ax.fill_between(x,y,where=(x>=5)&(x<=5.5),alpha=0.5,color='green')

# show plot & proportion
proportion=stats.norm.cdf(5.5,loc=5,scale=1.5)-stats.norm.cdf(5,loc=5,scale=1.5)
plt.show()
```

#### 2.4 What proportion of the gunny bags have a breaking strength is not between 3 and 7.5 kg per sq cm.?

```python
# joining all subplots
fig, ax=plt.subplots()

mu=5
sigma=1.5

# set lower & upper limits
lower=3
upper=7.5

# caliculate z-scores for lower & upper limit
z_lower = (lower-mu)/sigma
z_upper = (upper-mu)/sigma

# caliculate proportion using normal distribution function
lower_proportion = stats.norm.cdf(z_lower)
upper_proportion = stats.norm.cdf(z_upper)
proportion=lower_proportion+upper_proportion

# plot distributions
x=np.linspace(mu-3*sigma,mu+3*sigma,400)
y=norm.pdf(x,mu,sigma)
ax.plot(x,y)
ax.fill_between(x,y,where=(x<lower)|(x>upper),alpha=0.5,color='green')

# see title and lables
ax.set_xlabel('Breaking Strength')
ax.set_ylabel('probability Density')
ax.set_title('Normal Distribution with mean 5 and standard deviation 1.5')
plt.show()
```
